## Основные выводы

### Вывод 1

После первого слоя декодера запросы ранжируются лучше, чем сразу после начального отбора, но всё ещё хуже, чем в конце модели.

Это видно по:

$$
\mathrm{Overlap}(E1, L) > \mathrm{Overlap}(E0, L)
$$

и

$$
\mathrm{AP}(E1\text{-full}) < \mathrm{AP}(L\text{-full})
$$

Итог:
- первый слой декодера уже сильно улучшает порядок запросов;
- но финальные слои всё ещё полезны.

---

### Вывод 2

Ранняя остановка после первого слоя действительно ускоряет модель, но даёт заметную потерю качества.

Получилось примерно:

$$
T(L\text{-full}) \approx 1136.7\ \text{ms}
$$

$$
T(E1\text{-full}) \approx 997.8\ \text{ms}
$$

То есть выигрыш по времени:

$$
\Delta T \approx 138.9\ \text{ms}
$$

и ускорение:

$$
\mathrm{speedup} \approx 1.14\times
$$

Но при этом:

$$
\mathrm{AP}(L\text{-full}) = 0.597
$$

$$
\mathrm{AP}(E1\text{-full}) = 0.557
$$

То есть потеря:

$$
\Delta \mathrm{AP} \approx 0.040
$$

Итог:
- ранняя остановка реально полезна по времени;
- но она не бесплатна по качеству.

---

### Вывод 3

Позднее урезание числа запросов почти не портит качество, но почти не уменьшает вычисления.

Это видно из:

$$
\mathrm{AP}(L\text{-}k75) \approx \mathrm{AP}(L\text{-full})
$$

$$
\mathrm{AP}(L\text{-}k50) \approx \mathrm{AP}(L\text{-full})
$$

Но время почти одинаковое:

$$
T(L\text{-}k75) \approx T(L\text{-full}), \qquad
T(L\text{-}k50) \approx T(L\text{-full})
$$

Итог:
- в конце много запросов уже лишние как итоговые предсказания;
- но удалять их в конце почти бесполезно для ускорения, потому что вся тяжёлая обработка уже выполнена.
